## Let's do some dictionary practice


```python
# Dictionary Practice 
# Run this cell to create 4 lists (companies, name, price, volume)


companies=['CVX','AXP','MRK','IBM','DIS']
name = ['Chevron Corporation', 'American Express', 'Merck & Co., Inc.',
        'International Business Machines Corporation', 'The Walt Disney Company']
price = [110.24,135.6,82.46,151.1,139.14]
volume = [4869064,1888892,8502056,3233147,7872913]



```

```python
"""

Complete the function below.

This function uses the 4 lists defined above (companies, name, price, volume), and returns
a list of dicts where each dict has the following format:

{'company':'CVX', 'name': 'Chevron Corporation', 'price':110.24, 'volume':4869064}

"""

def CreateDowJonesDict():
    """
    name: CreateDowJonesDict
    input: no input (void)
    return: a list of dicts
    """
    places = []
    # iteration by position
    for i in range(0, len(companies)):
        co_place = {}
        co_place["company"] = companies[i]
        co_place["name"] = name[i]
        co_place["price"] = price[i]
        co_place["volume"] = volume[i]
        
        places.append(co_place)
        
    return places
        
```

```python
"""

Call CreateDowJonesDict() and save the return value to a variable called 'companylist'

"""
companylist = CreateDowJonesDict()
```

```python
companylist
```

```python
"""
1. Write a loop to print all company-name tuples using the
list of dicts created through calling CreateDowJonesDict()
Each tuple should have the following format: ('CVX', 'Chevron Corporation')

"""

for i in companylist:
    co = i["company"]
    name = i["name"] 
    co_name = (co, name)
   
    print(co_name)

    # prof: 
    # for i in companylist:
    # print((i["company"], i["name"]))
    
    


```

```python
"""
2. Write a loop to iterate over the Chevron dictionary and print the key names
Again, you should use the same list of dicts created above
If your list of dicts is called companylist, you can use companylist[0] to access to the Chevron dictionary
"""
# iterate over key
for element in companylist[0]:
    print(element)
    


```

```python
"""
3. Write a loop to iterate over the Chevron dictionary and print the (key, value) tuples
"""
# iterate over items
for element in companylist[0].items():
    print(element)
```

```python
"""
4. Write a loop to iterate over the Chevron dictionary and print the values
"""

# iterate over value

for element in companylist[0].values():
    print(element)








```

```python
"""
5. Use list's sorted function to print companylist sorted in alphabetical order
based on company ticker

https://www.programiz.com/python-programming/methods/built-in/sorted

"""

def company_tic(dict):
    return dict["company"]
# sort by company
sorted_com = sorted(companylist, key = company_tic)

print(sorted_com)

```

```python
"""
6. Use list's sorted function to print companylist sorted by volume in descending order
"""

def volume(dict):
    return dict["volume"]
# descending order by using reverse = true
sorted_vol = sorted(companylist, key = volume, reverse = True)

print(sorted_vol)


```

```python
"""
This cell imports the pandas package and gives it a nickname (pd). So after running this cell you will type pd whenever
you need to use pandas
"""
import pandas as pd
```

```python
"""

This cell reads in the file "starbucks_data.csv" into python as a DataFrame, which is a typical format to store data

"""
starbucks = pd.read_csv("starbucks_data.csv")
```

```python
print(starbucks['Calories'].mean())
print(starbucks['Calories'].sum())
print(starbucks['Calories'].max())
print(starbucks['Calories'].min())
print(starbucks['Calories'].std())
```

```python
'''
To access more than one column at a time, use a list as input
The return value is itself a DataFrame
'''
starbucks[['Beverage','Total Fat (g)']]
```

```python
'''

You can check the Truth value of a logical test for an entire column

'''
starbucks['Calories']>300
```

```python
'''

You can use the truth values to select rows that meet your criteria (only the "True" rows are returned)
The returned value is a DataFrame.

'''

starbucks[starbucks['Calories']>300]
```

```python
'''

You can easily access specified columns of a subset of DataFrame, because the subset is itself a DataFrame, so
all DataFrame functions and operations will work.

'''
starbucks[starbucks['Calories']>300][['Beverage','Calories']]
```

```python
"""
You can specify multiple conditions to filter rows. Remember to use parentheses to mark different conditions.

"""

starbucks[(starbucks['Calories']>300) & (starbucks['Calories']<350)][['Beverage','Calories']]
```

```python
"""

You can easily sort based on one or more columns

Below sort rows based on Calories.
"""

starbucks.sort_values('Calories')
```

```python
"""

Below sort rows based on Calories and then Total Fat (g).

"""
starbucks.sort_values(['Calories','Total Fat (g)'])
```
